```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'House size':[500,800,1000,1200,1500,1800,2000,2300],
    'price':[15,22,27,32,40,48,55,62]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

y = mx+b

x = input variable
y = target variable
m = slope
b = intercept

```python
x = df['House size']
y = df['price']
```

```python
df['x**2'] = x**2
```

```python
df['xy'] = x*y
```

```python
df
```

```python
print(sum(df['House size']))
print(sum(df['price']))
print(sum(df['x**2']))
print(sum(df['xy']))
```

```python
n = len(df)
n
```

m=n(sum(xy)-(sum(x))(sum(y))) / n(sum(x^2) - (sum(x))^2)

```python
m_numerator = n*(sum(df['xy'])) - sum(df['House size'])*(sum(df['price']) )
m_numerator
```

```python
m_denominator = n*(sum(df['x**2'])) - sum(df['House size'])**2
m_denominator
```

slope

```python
m = m_numerator / m_denominator
m
```

b = y(mean) - m*x(mean)

```python
b = (df['price'].mean()) - (m*df['House size'].mean())
b
```

```python
l = m*0+b
l
```

```python
df['predicted'] = m*x+b
df
```

```python
df['error'] = df['price']-df['predicted']
```

```python
df['squared error'] = df['error']**2
```

```python
df
```

```python
df['squared error'].mean()
```

```python
s = sum(abs((df['error'])))
s/8
```

```python
df['sum_of_total_variation'] = (df['price'] - df['price'].mean())**2
```
